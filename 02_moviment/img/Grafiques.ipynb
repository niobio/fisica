```python
#!/usr/bin/env python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
t = np.arange(1.0, 10, 0.01)
v = 2+0*t

fig1, ax1 = plt.subplots()

#ax1.fill_between(t, 0, v, hatch='//', facecolor= 'white')
ax1.fill_between(t, 0, v, facecolor= 'lightgrey', edgecolor='lightgrey')
ax1.set_xlim(0,10)
ax1.set_ylim(0,5)
ax1.set_ylabel('velocitat (m/s)')
ax1.set_xlabel('temps (s)')
ax1.grid()
ax1.plot(t,v, color='red', linewidth=2)
fig1.savefig('velocitat1.png')
```

```python
x=5+2*(t-1)

fig2, ax2 = plt.subplots()

#ax2.fill_between(t, 0, x, hatch='+', facecolor= 'white')
#ax2.fill_between(t, 0, x, facecolor= 'lightgrey', edgecolor='lightgrey')
ax2.set_xlim(0,10)
ax2.set_ylim(0,25)
#ax2.set_ylim(0,5)
ax2.set_ylabel(u'posició (m)')
ax2.set_xlabel('temps (s)')
ax2.grid()
ax2.plot(t,x, color='red', linewidth=2)
fig2.savefig('posicio1.png')
```
