```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

t = np.linspace(0,4,200)
v = 20*np.sin(2*np.pi*t)
f = plt.figure()
ax = f.add_subplot(1,1,1)
ax.set_xlabel('Temps (s)')
ax.set_ylabel('Tensi√≥ (V)')
ax.grid(True)
plt.plot(t,v)
f.savefig('tensio_alterna.png',dpi=300)
```

```python
pca = v**2
pcc = 200*np.ones(len(t))
```

```python

f2 = plt.figure()
ax2 = f2.add_subplot(1,1,1)
ax2.set_xlabel('Temps (s)')
ax2.set_ylabel(u'Pot√®ncia (W)')
plt.plot(t,pca,color='b', label='AC')
plt.plot(t, pcc,color='r',label='CC')
ax2.legend(loc='best')
ax2.fill_between(t, pca, facecolor='yellow', interpolate=True)
ax2.fill_between(t, pca, pcc, where=pca <= pcc, facecolor='cyan', interpolate=True)
f2.savefig('potencia.png',dpi=300)
```
